```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def sentence_similarity(sentence1, sentence2):
    # Tokenize and convert sentences to lowercase
    sentence1 = sentence1.lower()
    sentence2 = sentence2.lower()

    # Use CountVectorizer to convert sentences to vectors
    vectorizer = CountVectorizer().fit_transform([sentence1, sentence2])
    vectors = vectorizer.toarray()

    # Calculate cosine similarity between the vectors
    similarity = cosine_similarity(vectors)

    # The similarity matrix is a 2x2 matrix, and we're interested in the [0, 1] or [1, 0] element
    similarity_score = similarity[0, 1]

    return similarity_score

# Example usage:
sentence1 = "The quick brown fox jumps over the lazy dog."
sentence2 = "A lazy dog is jumped over by a quick brown fox."

similarity_score = sentence_similarity(sentence1, sentence2)
print(f"Cosine Similarity: {similarity_score}")
```

```python
from transformers import AutoTokenizer, AutoModel
import torch
import numpy as np

def collect_prompts(tool_name):
    # Your existing code to collect prompts
    if tool_name == 'observe point tool':
        return [
            "Use observe point tool to monitor website traffic",
            "How can I analyze data using observe point?",
            "observe point tool traffic analysis"
        ]
    elif tool_name == 'adobe analytics':
        return [
            "Analyze website performance with adobe analytics",
            "adobe analytics conversion rate report",
            "How to use adobe analytics for user behavior analysis?"
        ]
    elif tool_name == 'google analytics':
        return [
            "Google analytics dashboard overview",
            "Track website traffic using google analytics",
            "How to set up goals in google analytics?"
        ]
    else:
        return []  # Unknown tool, return an empty list


def sentence_similarity_bert(sentence1, sentence2, model, tokenizer):
    inputs = tokenizer([sentence1, sentence2], return_tensors="pt", padding=True, truncation=True)
    with torch.no_grad():
        outputs = model(**inputs)
    embeddings = outputs.pooler_output
    similarity = np.dot(embeddings[0], embeddings[1]) / (np.linalg.norm(embeddings[0]) * np.linalg.norm(embeddings[1]))
    return similarity

def determine_tool_bert(query, model, tokenizer):
    tool_names = ['observe point tool', 'adobe analytics', 'google analytics']

    tool_scores = {}

    for tool_name in tool_names:
        prompts = collect_prompts(tool_name)
        similarities = [sentence_similarity_bert(query, prompt, model, tokenizer) for prompt in prompts]
        avg_similarity = sum(similarities) / len(similarities) if len(similarities) > 0 else 0
        tool_scores[tool_name] = avg_similarity

    determined_tool = max(tool_scores, key=tool_scores.get)

    print('Tool Scores:', tool_scores)
    print('Determined Tool:', determined_tool)
    print('Average Similarity Score:', tool_scores[determined_tool])

    return determined_tool

# Download a pre-trained BERT model from Hugging Face
# For example, "bert-base-uncased" is a commonly used one
model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

query = "I want to use google analytics to monitor my website traffic"
tool = determine_tool_bert(query, model, tokenizer)
print(f"The user is trying to use {tool}")
```
